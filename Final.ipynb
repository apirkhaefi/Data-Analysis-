```python
data["Sleep Disorder"].fillna("None",inplace=True)
data
```

```python
#data["Gender"]=data["Gender"].astype(str)
#data["Gender"]=data["Gender"].str.strip().str.title()
#data.drop("Gender_Num",axis=1,inplace=True)
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})

data.head()
```

```python
data["Occupation"].unique()
```

```python
occupation_map = {
    'Doctor': 'Medical',
    'Nurse': 'Medical',
    'Software Engineer': 'High_Tech',
    'Engineer': 'High_Tech',
    'Scientist': 'High_Tech',
    'Salesperson': 'Sales',
    'Sales Representative': 'Sales',
    'Manager': 'Sales',
    'Teacher': 'Education',
    'Lawyer': 'Legal',
    'Accountant':'Sales'
}
```

```python
data['Occupation_Grouped'] = data['Occupation'].map(occupation_map)
data.head()
```

```python
data.drop("Occupation",axis=1,inplace=True)
data.head()
```

```python
data = pd.get_dummies(data, columns=['Occupation_Grouped'], drop_first=True)
data
```

```python
data['Has_Disorder'] = data['Sleep Disorder'].map({'None': 0, 'Insomnia': 1, 'Sleep Apnea': 1})
data
```

```python
data.drop("Sleep Disorder",axis=1,inplace=True)
data
```

```python
data["Blood Pressure"].unique()
```

```python
data[['Systolic_BP', 'Diastolic_BP']] = data['Blood Pressure'].str.split('/', expand=True).astype(int)
data
```

```python
data.drop("Blood Pressure",axis=1,inplace=True)
data
```

```python
data.drop("Person ID",axis=1,inplace=True)
data
```

```python
data["BMI Category"].unique()
```

```python
data = pd.get_dummies(data, columns=['BMI Category'], drop_first=True)
data
```

```python
correlation=data.corr()
plt.figure(figsize=(15,8))
sb.heatmap(correlation,annot=True,cmap="viridis",fmt="0.2f",vmin=-1,vmax=1)
plt.title("correlation")
plt.show()
```

```python
corr = data.corr().abs()  # قدر مطلق برای تمرکز بر شدت رابطه
corr_pairs = corr.unstack().sort_values(ascending=False)
# فیلتر: فقط بین ویژگی‌های متفاوت (نه خودش با خودش) و بالای 0.5
strong_corr = corr_pairs[(corr_pairs < 1) & (corr_pairs > 0.5)]
strong_corr
```

```python
from scipy.stats import pearsonr,spearmanr
x=data['Sleep Duration']
y=data['Quality of Sleep']
p_coeff,p_value=pearsonr(x,y)
print(f"Pearson correlation: {p_coeff:.3f}, p-value: {p_value:.4f}")
P_coeff,p_value=spearmanr(x,y)
print(f"Spearman correlation: {p_coeff:.3f}, p-value: {p_value:.4f}")
```

```python
from scipy import stats 
t_stats,p_vaalue=stats.ttest_1samp(data["Sleep Duration"],7.5)
print("t_statistic",t_stats)
print("p_value",p_value)
if p_value<0.05 :
    print("reject")
else:
    print("accept")
```
